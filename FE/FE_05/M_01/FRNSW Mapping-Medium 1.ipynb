```python
from bs4 import BeautifulSoup
import requests
```

```python
zones = ['ME1', 'ME2', 'ME3', 'MW1', 'MW2', 'MW3', 'MN1', 'MN2', 'MN3', 'MS1', \
         'MS2', 'MS3', 'RW1', 'RW2', 'RW3', 'RN1', 'RN2', 'RN3', 'RS1', 'RS2', 'RS3']
```

```python
url_base = 'https://www.fire.nsw.gov.au/page.php?id=9231&zone='
```

```python
sta = []
add = []

for zone in zones:
    url = url_base + zone
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    tags = soup.find_all('div', class_ = "card-content")
    station = []
    address = []
    for tag in tags:
        station.append(tag.span.text)
        address.append(tag.p.text)
    print(station)
    print(address)
    sta.append(station)
    add.append(address)
print(sta)
print(add)
```
