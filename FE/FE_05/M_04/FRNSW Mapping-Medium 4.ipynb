```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib
import urllib.parse
from geopy.geocoders import Nominatim
import json
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv(r'FRNSW_lat_lng.csv')
```

```python
df
```

```python
m = folium.Map(location = [-33.8, 151.2], tiles ='OpenStreetMap', zoom_start=6)
for i, row in df.iterrows():
    lat = df.at[i, 'lat']
    lng = df.at[i, 'lng']
    #add = df.at[i, 'Address']
    #stn = df.at[i, 'Station No.']
    
    popup = 'Station : ' + str(df.at[i, 'Stations']) + '<br>' + \
            '<br>' +'Address: ' + df.at[i, 'Address']

    #popup = folium.Popup(iframe, max_width=200)
    folium.Marker(location = [lat, lng], popup= popup, icon = folium.Icon(color='blue')).add_to(m)

m.save('FRNSW_Locations.html')
```

```python
m = folium.Map(location = [-33.8, 151.2], tiles = 'OpenStreetMap', zoom_start=6)
markerCluster = MarkerCluster().add_to(m)
for i, row in df.iterrows():
    lat = df.at[i, 'lat']
    lng = df.at[i, 'lng']
    #add = df.at[i, 'Address']
    #stn = df.at[i, 'Station No.']
    
    popup = 'Station : ' + str(df.at[i, 'Stations']) + '<br>' + '<br>' +'Address: ' + df.at[i, 'Address']
    iframe = folium.IFrame(popup, width=200,  height=150)

    #popup = folium.Popup(iframe, max_width=200)
    folium.Marker(location = [lat, lng], popup= popup, icon = folium.Icon(color='blue', icon='fire')).add_to(markerCluster)

m.save('FRNSW_Locations3.html')
```

```python
m = folium.Map(location = [-33.8, 151.2], tiles = 'OpenStreetMap', zoom_start=6)
geojson = r'australian-states-NSW.json'
markerCluster = MarkerCluster().add_to(m)
for i, row in df.iterrows():
    lat = df.at[i, 'lat']
    lng = df.at[i, 'lng']
    #add = df.at[i, 'Address']
    #stn = df.at[i, 'Station No.']
    
    popup = 'Station : ' + str(df.at[i, 'Stations']) + '<br>' + '<br>' +'Address: ' + df.at[i, 'Address']
    iframe = folium.IFrame(popup, width=200,  height=150)

    #popup = folium.Popup(iframe, max_width=200)
    folium.Marker(location = [lat, lng], popup= popup, icon = folium.Icon(color='blue', icon='fire')).add_to(markerCluster)

style = {'fillColor': '#00000000', 'color': '#00000000'}
folium.GeoJson(
    geojson, 
    name='geojson',
    style_function = lambda x: {
        'color': 'blue',
        'weight': 1,
        "opacity":1,
        'fillOpacity': 0,
        'interactive':False
    }).add_to(m)
# folium.GeoJsonTooltip(fields=['STATE_NAME']).add_to(g)
m.save('frnsw_location4.html')
```
