## Import relevant libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data to a variable df 
df = pd.read_csv(r'C:\Users\OneDrive\JabirJamal.com\Article1\Model1\Model1_hrr.csv')
```

```python
df.head()
```

## Change column name

```python
# create a copy of the dataframe
df1 = df.copy()

# copy row 0 of df to header of df1
df1.columns=df.iloc[0]
df1.head()
```

## Remove row 0

```python
df1=df1[1:].reset_index(drop=True)
df1.head()
```

## Change datatype

```python
df1 = df1.astype('float')
df1.head()
```

## Generate HRR graph

```python
# change working directory to the location where you would like to save the image
os.chdir(r'C:\Users\OneDrive\JabirJamal.com\Article1\Model1')

# extract x and y columns to generate the graph
df1.plot(x='Time', y='HRR', title = 'Model 1: Heat Release Rate (kW)', figsize=(7,5)) 

# label x and y axis
plt.xlabel('Time (s)')
plt.ylabel('HRR (kW)')

# specify limits for x and y axis
plt.xlim(0,300)
plt.ylim(0,300)

# specify file name for the image and generate the graph
plt.savefig('Model1_HRR.png', bbox_inches='tight')
plt.show()
```
